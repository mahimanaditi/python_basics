```python
"""
ðŸ“˜ Python Basics for Data Science
Author: Aditi Mahiman
GitHub: https://github.com/mahimanaditi
"""

# ---------------------------------------
#  Variables and Data Types
# ---------------------------------------

# Integers, floats, strings, booleans
x = 10         # Integer
pi = 3.14      # Float
name = "Aditi" # String
is_valid = True # Boolean

# type() function
print(type(x), type(pi), type(name), type(is_valid))


# ---------------------------------------
#  Operators
# ---------------------------------------

# Arithmetic operators
a = 10
b = 3
print("Addition:", a + b)
print("Power:", a ** b)

# Logical operators
print("AND:", a > 5 and b < 5)


# ---------------------------------------
#  Flow Control
# ---------------------------------------

# If-Else
age = 20
if age >= 18:
    print("Eligible to vote")
else:
    print("Not eligible")


# ---------------------------------------
#  Loops
# ---------------------------------------

# For loop
for i in range(3):
    print("For loop:", i)

# While loop
count = 0
while count < 2:
    print("While loop:", count)
    count += 1


# ---------------------------------------
#  Collections
# ---------------------------------------

# Strings
text = "Data Science"
print(text.upper())

# List
fruits = ["apple", "banana"]
fruits.append("mango")

# Tuple
colors = ("red", "green")

# Dictionary
student = {"name": "Aditi", "age": 26}


# ---------------------------------------
#  List Comprehension
# ---------------------------------------

squares = [x**2 for x in range(5)]
print("Squares:", squares)


# ---------------------------------------
#  Functions
# ---------------------------------------

# User-defined function
def greet(name):
    """Returns a greeting message."""
    return "Hello " + name

# Default arguments
def power(base, exponent=2):
    return base ** exponent

# Lambda function
square = lambda x: x**2
print(square(5))
```
